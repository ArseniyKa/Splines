```python
# !pip install pandas
# !pip install odfpy
import pandas as pd
import numpy as np
import matplotlib
from numpy import linalg as LA

matplotlib.use('TkAgg')
import matplotlib.pyplot as plt

def Fitting(x, y, polinom_size):
        vandermond_tr = np.zeros((polinom_size,x.size))
        
        for i in range(polinom_size):
              vandermond_tr[i] = x**i
        
        vandermond = vandermond_tr.transpose()
        
        XX = vandermond_tr.dot(vandermond)
        XX_inv = np.linalg.inv(XX)
        XY = vandermond_tr.dot(y)
        coeff_vector = XX_inv.dot(XY)

        return coeff_vector
    
def GaussMethod(x, y, Xo, Func_value, polinom_size):
    vandermond_tr = np.zeros((polinom_size,x.size))
   
    xo = Xo
    yo = Func_value
    lagrange_vec = np.zeros((polinom_size,1))
    condition1 = np.zeros(polinom_size+1)
        
    for i in range(polinom_size):
        vandermond_tr[i] = x**i
        lagrange_vec[i] = -0.5 * xo**i
        condition1[i] = xo**i
    
    vandermond = vandermond_tr.transpose()
    condition1 = condition1.transpose()
        
    XX = vandermond_tr.dot(vandermond)
    XY = vandermond_tr.dot(y)
    XY = np.append(XY, yo)
#     print(XY)
    print(condition1)


#     A_new = np.concatenate((XX, lagrange_vec.T), axis=1)
    A = np.hstack((XX,lagrange_vec))
    A = np.vstack((A, condition1))
    print(lagrange_vec)
    print("------")
    print(A)
    

    coeff_vector = LA.solve(A, XY)
    return coeff_vector

###################################################
pd.set_option("display.max.columns", None)
df1 = pd.read_excel("sample.xlsx", sheet_name='list1')

discrete_size= 1000
step = 1./discrete_size

#######
# print(df1)
arr1 = df1.to_numpy()
in1=arr1[:,0]
out1=arr1[:,1]

x=in1[::-1]
y = out1[::-1]
######################################
mg=0.00001
ln_y = np.log(y)
ln_x = np.log(1-x**mg)

begin = 0
end = 1
coeff_matrix = GaussMethod(ln_x,ln_y, ln_x[0], ln_y[0], 3)
# coeff_matrix = Fitting(ln_x,ln_y, 2)
temp = np.linspace(np.min(ln_x), np.max(ln_x), discrete_size)
Y = coeff_matrix[0] + coeff_matrix[1] * temp

# coeff_matrix =np.zeros((number,4))
# splines = np.zeros((number, discrete_size ))
# first_derivative = np.zeros((number, discrete_size))
# second_derivative = np.zeros((number,
# discrete_size))

    
```
